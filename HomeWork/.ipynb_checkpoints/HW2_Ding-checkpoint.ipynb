In this question, you will explore the application of Lasso and Ridge regression using sklearn package in Python. 1) Use sklearn.linear_model.Lasso and sklearn.linear_model.Ridge classes to do a [5-fold cross validation](http://scikit-learn.org/stable/auto_examples/exercises/plot_cv_diabetes.html#example-exercises-plot-cv-diabetes-py) using sklearn's [KFold](http://scikit-learn.org/stable/modules/generated/sklearn.cross_validation.KFold.html). For the sweep of the regularization parameter, we will look at a grid of values ranging from $\lambda = 10^{10}$ to $\lambda = 10^{-2}$. In Python, you can consider this range of values as follows:

      alphas =  10**np.linspace(10,-2,100)*0.5

  Report the best chosen $\lambda$ based on cross validation. The cross validation should happen on your training data using  average MSE as the scoring metric. (5pts)

2) Run ridge and lasso for all of the alphas specified above (on training data), and plot the coefficients learned for each of them - there should be one plot each for lasso and ridge, so a total of two plots; the plots for different features for a method should be on the same plot (e.g. Fig 6.6 of JW). What do you qualitatively observe when value of the regularization parameter is changed? (4pts)

3) Run least squares regression, ridge, and lasso on the training data. For ridge and lasso, use only the best regularization parameter. Report the prediction error (MSE) on the test data for each. (3pts)

4) Run lasso again with cross validation using [sklearn.linear_model.LassoCV](http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LassoCV.html). Set the cross validation parameters as follows:

    LassoCV(alphas=None, cv=10, max_iter=10000)

Report the best $\lambda$ based on cross validation. Run lasso on the training data using the best $\lambda$ and report the coefficeints for 55 variables. What do you observe from these coefficients? (3pts)

```python
# (1)
%matplotlib inline
from __future__ import print_function
print(__doc__)

import numpy as np
import matplotlib.pyplot as plt
import sys

from sklearn import datasets, linear_model
from sklearn import datasets
from sklearn.linear_model import LassoCV
from sklearn.linear_model import Lasso
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge

from sklearn import datasets
diabetes = datasets.load_diabetes()
X = diabetes.data
y = diabetes.target

from sklearn.preprocessing import PolynomialFeatures
X = PolynomialFeatures(2, interaction_only=True, include_bias=False).fit_transform(X)

from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.33, random_state=20)

k_fold = KFold(5)
min_Lasso_MSE = sys.float_info.max
best_Lasso_alpha = 0.1
min_Ridge_MSE = sys.float_info.max
best_Ridge_alpha = 0.1
alphas =  10**np.linspace(10,-2,100)*0.5
lasso = Lasso()
ridge = Ridge()
lasso_coefs = []
ridge_coefs = []
for alpha in alphas:
    lasso.set_params(alpha=alpha, random_state=20)
    ridge.set_params(alpha=alpha)
    lasso_sum = 0
    ridge_sum = 0
    lasso.fit(X_train, y_train)
    lasso_coefs.append(lasso.coef_)
    ridge.fit(X_train, y_train)
    ridge_coefs.append(ridge.coef_)
    for k, (train, test) in enumerate(k_fold.split(X_train, y_train)):
        lasso.fit(X_train[train], y_train[train])
        ridge.fit(X_train[train], y_train[train])
        lasso_sum += mean_squared_error(y_train[test], lasso.predict(X_train[test]))
        ridge_sum += mean_squared_error(y_train[test], ridge.predict(X_train[test]))
    if lasso_sum/5 < min_Lasso_MSE:
        min_Lasso_MSE = lasso_sum/5
        best_Lasso_alpha = alpha
    if ridge_sum/5 < min_Ridge_MSE:
        min_Ridge_MSE = ridge_sum/5
        best_Ridge_alpha = alpha
        
print("Lasso_alpha: {0:.5f}, Lasso_MSE: {1:.5f}".format(best_Lasso_alpha, min_Lasso_MSE))
print("Ridge_alpha: {0:.5f}, Ridge_MSE: {1:.5f}".format(best_Ridge_alpha, min_Ridge_MSE))
```

```python
# (2)
ax = plt.gca()

ax.plot(alphas, lasso_coefs)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Lasso coefficients as a function of the regularization')
plt.axis('tight')
plt.show()

ax = plt.gca()

ax.plot(alphas, ridge_coefs)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Ridge coefficients as a function of the regularization')
plt.axis('tight')
plt.show()
print()
print("\nConclusion: As alpha(lambda) increases, the model complexity decreases until all coefficients become zero which means the model becomes a constant.")
```

```python
# (3)
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)
print("MSE of LSR: %.2f"
      % np.mean((regr.predict(X_test) - y_test) ** 2))

lasso.set_params(alpha=best_Lasso_alpha, random_state=20)
lasso.fit(X_train, y_train)
print("MSE of Lasso: %.2f"
      % np.mean((lasso.predict(X_test) - y_test) ** 2))

ridge.set_params(alpha=best_Ridge_alpha)
ridge.fit(X_train, y_train)
print("MSE of LSR: %.2f"
      % np.mean((ridge.predict(X_test) - y_test) ** 2))
```

```python
# (4)
lassoCV = LassoCV(alphas=None, cv=10, max_iter=10000)
lassoCV.fit(X_train, y_train)
lasso.set_params(alpha=lassoCV.alpha_, random_state=20)
lasso.fit(X_train, y_train)
print("best lambda: %.3f" %lassoCV.alpha_)
print("\n coefficients:")
print(lasso.coef_)
print("\nConclusion: lasso can eliminate the irrelevant factors")
```

# Question 2: Multi-layer Perceptron Regressor (15 points)

In this question, you will explore the application of Multi-layer Perceptron (MLP) regression using sklearn package in Python. We will use the Boston house-prices dataset for this problem http://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_boston.html.

Following code will load and split the data into training and test set using [train_test_split](http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html) with **random state 42** and **test_size = 0.33**.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import (train_test_split,KFold)
from sklearn.metrics import mean_squared_error
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets
data = datasets.load_boston()
feature_cols = data.feature_names
X = pd.DataFrame(data.data, columns = feature_cols)
y = pd.Series(data.target)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.33, random_state=42)
```

One more thing to use in this problem is [StandardScaler](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html).    Report the average Root Mean Squared Error (RMSE) value based on your 5-fold cross validation for each model: (i), (ii), and (iii) (5pts)



```python
import pandas as pd
import numpy as np
from sklearn.model_selection import (train_test_split,KFold)
from sklearn.metrics import mean_squared_error
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets
data = datasets.load_boston()
feature_cols = data.feature_names
X = pd.DataFrame(data.data, columns = feature_cols)
y = pd.Series(data.target)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.33, random_state=42)

print('1)')
# Convert type to ndarray
X_train = X_train.as_matrix()
X_test = X_test.as_matrix()
y_train = y_train.as_matrix()
y_test = y_test.as_matrix()
X_scaler = StandardScaler()
X_train = X_scaler.fit_transform(X_train)
X_test = X_scaler.transform(X_test)
k_fold = KFold(5)
mlp_a = MLPRegressor(hidden_layer_sizes=(2,), activation='tanh', solver='sgd', learning_rate='constant', random_state=42,
batch_size=40, learning_rate_init = 0.005)
rmse_sum = 0
for k, (train, validate) in enumerate(k_fold.split(X_train, y_train)):
    mlp_a.fit(X_train[train], y_train[train])
    rmse_sum += np.sqrt(mean_squared_error(y_train[validate], mlp_a.predict(X_train[validate])))
print('(a). Root Mean of Squared Errors is (hidden layer size is 2):', rmse_sum / 5)

mlp_b = MLPRegressor(hidden_layer_sizes=(8,), activation='tanh', solver='sgd', learning_rate='constant', random_state=42,
batch_size=40, learning_rate_init = 0.005)
rmse_sum = 0
for k, (train, validate) in enumerate(k_fold.split(X_train, y_train)):
    mlp_b.fit(X_train[train], y_train[train])
    rmse_sum += np.sqrt(mean_squared_error(y_train[validate], mlp_b.predict(X_train[validate])))
print('(b). Root Mean of Squared Errors is (hidden layer size is 8):', rmse_sum / 5)

mlp_c = MLPRegressor(hidden_layer_sizes=(15,), activation='tanh', solver='sgd', learning_rate='constant', random_state=42,
batch_size=40, learning_rate_init = 0.005)
rmse_sum = 0
for k, (train, validate) in enumerate(k_fold.split(X_train, y_train)):
    mlp_c.fit(X_train[train], y_train[train])
    rmse_sum += np.sqrt(mean_squared_error(y_train[validate], mlp_c.predict(X_train[validate])))
print('(c). Root Mean of Squared Errors is (hidden layer size is 15):', rmse_sum / 5)
print()

print('2)')
mlp_a.fit(X_train, y_train)
print('(a). RMSE on training data is (hidden layer size is 2):',
      np.sqrt(mean_squared_error(y_train, mlp_a.predict(X_train))))
print('     RMSE on testing data is (hidden layer size is 2):',
      np.sqrt(mean_squared_error(y_test, mlp_a.predict(X_test))))

mlp_b.fit(X_train, y_train)
print('(b). RMSE on training data is (hidden layer size is 8):',
      np.sqrt(mean_squared_error(y_train, mlp_b.predict(X_train))))
print('     RMSE on testing data is (hidden layer size is 8):',
      np.sqrt(mean_squared_error(y_test, mlp_b.predict(X_test))))

mlp_c.fit(X_train, y_train)
print('(c). RMSE on training data is (hidden layer size is 15):',
      np.sqrt(mean_squared_error(y_train, mlp_c.predict(X_train))))
print('     RMSE on testing data is (hidden layer size is 15):',
      np.sqrt(mean_squared_error(y_test, mlp_c.predict(X_test))))
print('The result suggests that as the hidden layer size increases, the MSEs on both training and test set decrease.')
print()

print('3)')
plt.plot(mlp_a.loss_curve_)
plt.plot(mlp_b.loss_curve_)
plt.plot(mlp_c.loss_curve_)
plt.show()

print('4)')
mlp = MLPRegressor(activation = 'tanh', solver = 'sgd', learning_rate='constant', random_state=42,
batch_size=40, hidden_layer_sizes = (15,), learning_rate_init = 0.005)
rmse_sum = 0
for k, (train, validate) in enumerate(k_fold.split(X_train, y_train)):
    mlp.fit(X_train[train], y_train[train])
    rmse_sum += np.sqrt(mean_squared_error(y_train[validate], mlp.predict(X_train[validate])))
print('(i). Root Mean of Squared Errors is (learning_rate_init = 0.005):', rmse_sum / 5)

mlp.set_params(learning_rate_init=0.01)
rmse_sum = 0
for k, (train, validate) in enumerate(k_fold.split(X_train, y_train)):
    mlp.fit(X_train[train], y_train[train])
    rmse_sum += np.sqrt(mean_squared_error(y_train[validate], mlp.predict(X_train[validate])))
print('(ii). Root Mean of Squared Errors is (learning_rate_init = 0.01):', rmse_sum / 5)

mlp.set_params(learning_rate_init=1)
rmse_sum = 0
for k, (train, validate) in enumerate(k_fold.split(X_train, y_train)):
    mlp.fit(X_train[train], y_train[train])
    rmse_sum += np.sqrt(mean_squared_error(y_train[validate], mlp.predict(X_train[validate])))
print('(iii). Root Mean of Squared Errors is (learning_rate_init = 1):', rmse_sum / 5)
```

# Question 6 - Variable Selection Property of LASSO (5pts)

Explain in your own words (max 2 paras) what the diamond, circular and elliptical shapes in Fig 6.7 of JW (page 222) mean, and how this figure tries to explain why lasso tends to drive the values of some parameters to zero more aggressively as compared to ridge regression.

The diamond and circular shapes mean the constraints on coefficients for the lasso and ridge regression respectively. The elliptical shape indicates coefficients with the same RSS. The coefficients for the lasso and ridge regression are given by the first point at which an ellipse contacts the diamond or circular region. Intuitively, it is more likely for the diamond to contact with the elliptical at the axis where the coefficients become zero since the diamond has sharp ridges compared to the circle shape.
